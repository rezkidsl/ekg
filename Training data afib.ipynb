```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df_afib = pd.read_csv('D:/Data Machinelearning/AFIB.csv', skiprows=[1]) #read csv file
df_normal = pd.read_csv('D:/Data Machinelearning/NORMAL.csv', skiprows=[1]) #read csv file   
df_afib.head() #show first 5 rows
```

```python
df_normal.head() #show first 5 rows
```

```python
df_afib.info() #show info
```

```python
df_afib.drop(df_afib.iloc[:, 8:16], inplace=True, axis=1) #drop column
df_normal.drop(df_normal.iloc[:, 8:16], inplace=True, axis=1) #drop column kalau TP sama PQ gk di drop tinggal  comand blok ini
```

```python
df_afib['label'] = ('AFIB') #add label
df_normal['label'] = ('NORMAL') #add label
```

```python
new_df=pd.concat([df_afib,df_normal],axis=0) #concatenate 2 dataframe
new_df.head() #show first 5 rows
```

```python
new_df=new_df.sample(frac=1).reset_index(drop=True) #shuffle dataframe
```

```python
new_df.head()
```

```python
X = new_df.drop(['label'], axis=1) #drop label
y = new_df['label'] #label
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier

models = {
    'RandomForestClassifier': RandomForestClassifier(),
    'AdaBoostClassifier':AdaBoostClassifier(),
    'GradientBoostingClassifier': GradientBoostingClassifier()
}
```

```python
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
import pickle
kf = KFold(n_splits=3,shuffle=True,random_state=42) #split data n split bisa di ganti sesuai kebutuhan
metric='accuracy'
for name, model in models.items():
    pipeline = Pipeline( steps=[('model', model)])
    i=0
    for train_index, test_index in kf.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        pipeline.fit(X_train, y_train)
        y_pred = pipeline.predict(X_test)
        print(name, metric, accuracy_score(y_test, y_pred))
        pickle.dump(pipeline, open('model_'+name+'_'+str(i)+'.pkl', 'wb'))
        i=i+1
```
